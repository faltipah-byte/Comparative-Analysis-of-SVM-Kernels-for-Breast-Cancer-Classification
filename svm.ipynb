Comparative-Analysis-of-SVM-Kernels-for-Breast-Cancer-Classification

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
#import dataset breast cancer
bc = load_breast_cancer()

data = pd.DataFrame(bc.data, columns=bc.feature_names)
data["Class"] = bc.target

print(data.head())
print(data.describe())
```

```python
#EDA
print(data.info())
#cek missing value
print(data.isnull().sum())
```

```python
#split data
from sklearn.model_selection import train_test_split

X = data.drop("Class", axis=1)
y = data["Class"]

# ambil 500 data untuk training
X_train = X.sample(n=500, random_state=123)
y_train = y.loc[X_train.index]

X_test = X.drop(X_train.index)
y_test = y.drop(X_train.index)

print(len(X_train), len(X_test))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
# buat model SVM dengan kernel RBF
model = SVC(kernel="rbf", C=1, gamma="scale", random_state=123)
model.fit(X_train, y_train) 
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
print(f"Accuracy: {acc:.2f}")

#model svm linear
model_linear = SVC(kernel="linear", C=1, gamma="scale", random_state=123)
model_linear.fit(X_train, y_train)  
y_pred_linear = model_linear.predict(X_test)
cm_linear = confusion_matrix(y_test, y_pred_linear)
acc_linear = accuracy_score(y_test, y_pred_linear)
print("Confusion Matrix (Linear Kernel):")
print(cm_linear)
print(f"Accuracy (Linear Kernel): {acc_linear:.2f}")

#model svm polynomial
model_poly = SVC(kernel="poly", C=1, gamma="scale", random_state=123)
model_poly.fit(X_train, y_train)        
y_pred_poly = model_poly.predict(X_test)
cm_poly = confusion_matrix(y_test, y_pred_poly)
acc_poly = accuracy_score(y_test, y_pred_poly)
print("Confusion Matrix (Polynomial Kernel):")
print(cm_poly)
print(f"Accuracy (Polynomial Kernel): {acc_poly:.2f}")

#model svm sigmoid
model_sigmoid = SVC(kernel="sigmoid", C=1, gamma="scale", random_state=123)
model_sigmoid.fit(X_train, y_train) 
y_pred_sigmoid = model_sigmoid.predict(X_test)
cm_sigmoid = confusion_matrix(y_test, y_pred_sigmoid)
acc_sigmoid = accuracy_score(y_test, y_pred_sigmoid)
print("Confusion Matrix (Sigmoid Kernel):")
print(cm_sigmoid)
print(f"Accuracy (Sigmoid Kernel): {acc_sigmoid:.2f}")

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

#evaluasi parameter dengan grid search dan cross validation
param_grid = {  
    "C": [0.1, 1, 10],
    "gamma": ["scale", "auto"],
    "kernel": ["rbf", "linear", "poly", "sigmoid"]
}
grid_search = GridSearchCV(SVC(random_state=123), param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
cm_best = confusion_matrix(y_test, y_pred_best)
acc_best = accuracy_score(y_test, y_pred_best)
print("Confusion Matrix (Best Model):")
print(cm_best)
print(f"Accuracy (Best Model): {acc_best:.2f}")
```
